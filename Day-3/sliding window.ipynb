## Sub Arry

### Step:1

```python
l = [1,8,9]
n = len(l)
for i in range(n):
    for j in range(n):
        print(i,j)
```

### Stpe:2

```python
l = [1,8,9]
n = len(l)
for i in range(n):
    for j in range(i,n):
        print(i,j)
```

### Step:3

```python
l = [1,8,9]
n = len(l)
ans = []
for i in range(n):
    for j in range(i,n):
        temp = []
        for k in range(i,j+1):
            temp.append(l[k])
        ans.append(temp)
print(ans)
```

```python
l = [5,9,1,8,7]
n = len(l)
sub_array_len = 2
ans = []
for i in range(n):
    for j in range(i,n):
        temp = []
        for k in range(i,j+1):
            temp.append(l[k])
        if len(temp) == sub_array_len:
            ans.append(sum(temp))
print(max(ans))
```
