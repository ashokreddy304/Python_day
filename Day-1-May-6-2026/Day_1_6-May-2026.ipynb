```python
n = int(input('Enter 3 digit number'))

original = n
reversed_n = 0

while n != 0:
    digit = n%10
    reversed_n = reversed_n*10 + digit
    n = n//10

print(f"Original: {original}")
print(f"Reversed: {reversed_n}")
```

```python
l = [1, 2, 3, 4, 5]
target = 5

for i in range(len(l)):
    for j in range(i+1, len(l)):
        if l[i] + l[j] == target:
            print(f"Pair found: [{l[i]}, {l[j]}]")
```

```python
l = [1, 2, 3, 4, 5]
target = 5

seen = {}

for num in l:
    complement = target - num
    if complement in seen:
        print(f"Pair found: [{complement}, {num}]")
    seen[num] = True
```

```python
l = [3, 1, 4, 1, 5, 9]

first = second = float('-inf')

for num in l:
    if num > first:
        second = first
        first = num
    elif num > second and num != first:
        second = num

print(second)  # 5
```

```python
l = [3, 1, 4, 1, 5, 9]
k = 3

used = set()

for _ in range(k):
    max_val = float('-inf')
    
    for num in l:
        if num not in used and num > max_val:
            max_val = num
    
    used.add(max_val)

print(max_val)  # 5
```
