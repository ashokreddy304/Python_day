```python
l = [1, 2, 3, 4, 5]
k = 2  # rotate by 2
print(l[k:] + l[:k])   # [3, 4, 5, 1, 2]
```

```python
l = [1,8,9]
n = len(l)
ans = []
for i in range(n):
    for j in range(i,n):
        temp = []
        for k in range(i,j+1):
            temp.append(l[k])
        ans.append(temp)
print(ans)
```

```python
l = [5,9,1,8,7]
ans = 0
sub_array_len = 5
n = len(l)
for  i in range(n):
    for j in range(i,n):
        temp = []
        tsum = 0
        for k in range(i,j+1):
            temp.append(l[k])
            tsum += l[k]
        if len(temp) == sub_array_len:
            ans = max(ans,tsum)
print(ans)
```
